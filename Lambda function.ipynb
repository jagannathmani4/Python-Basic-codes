```python
#add 10 with given number
add10 = lambda val : val + 10
num = int(input("Enter a number:"))
print(f"{num} + 10 = {add10(num)}")
```

```python
#multiply with 5
mul5 = lambda val : val * 5
num = int(input("Enter a number:"))
print(f"{num} * 5 = {mul5(num)}")
```

```python
#square of a number
square = lambda val : val * val
num = int(input("Enter a number:"))
print(f"square of {num} = {square(num)}")
```

```python
#check positive number
is_positive = lambda val : val > 0
num = int(input("Enter a number:"))
print(f"{num} is {'positive' if is_positive(num) else 'negative'}")
```

```python
#last char of string
last_char = lambda text : text[(len(text)-1)]
print(last_char("Saheb"))
```

```python
#check if string length more than 5
length = lambda text : "less than 5" if len(text) < 5 else ("more than 5" if len(text) > 5 else "of 5")
str1 = input("Enter a string:")
print(f"This string {str1} has length {length(str1)}")
```

```python
#max of two numbers
max_num = lambda a,b : "Both are same" if (a == b) else(a if a>b else b)
print(max_num(35,75))
```

```python
#convert a string into uppercase
to_upper = lambda str1 : str1.upper()
text = input("Enter any string:")
print(to_upper(text))
```

```python
#check if divisible by num or not
is_divisible = lambda num, div : "Divisible" if(num%div == 0) else "Not divisible"
num1 = int(input("Enter a number:"))
div1 = int(input("Enter the dividor number:"))
print(f"{num1} is {is_divisible(num1,div1)} by {div1}")
```

```python
#first element of a list
first_element = lambda list1 : list1[0]
print(first_element([10,54,64,75,84,94]))
```
