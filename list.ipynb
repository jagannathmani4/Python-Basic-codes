```python
list = [5,48,21,15,36,54,64,20,42]
print(list)
print(list[0])
print(list[0:4])
print(list[0:9:2])
```

```python
list = [5,48,21,15,36,54,64,20,42]
print(list[:9:])
print(list[:9:2])
```

```python
list = [5,48,21,15,36,54,64,20,42]
print(list[-1::-1])
print(list[-1::-2])
```
