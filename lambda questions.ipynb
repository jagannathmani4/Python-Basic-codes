```python
num = int(input("Enter a number:"))
sign = lambda val: "Positive" if val>0 else "Negative" if val < 0 else "Not positive not negative"
print(sign(num))
```

```python
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
minimum = lambda num1, num2: num1 if num1 < num2 else num2 if num2 < num1 else "Both are same"
print(minimum(num1,num2))
```

```python
text = input("Enter a string:")
find_a = lambda text: "Available" if "a" in text else "Not available"
print(find_a(text))
```

```python
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
sub = lambda num1, num2: num1 - num2
print(sub(num1,num2))
```

```python
list = []
n = int(input("Enter total no of list numbers: "))
for i in range(n):
    num = int(input("Enter a number:"))
    list.append(num)

sum_list = lambda list: sum(list)
print(sum_list(list))
```
