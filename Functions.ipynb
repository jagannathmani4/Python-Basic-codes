```python
#waf to print the length of a list
def len_list(list1):
    length = 0
    for items in list1:
        length += 1
    print(length)
```

```python
#waf to print the length of a list
#22
l1 = [10,15,24,75,16]
print(len_list(l1))
```

```python
#WAF to print the elements of a list in a single line
def show_list_items(list1):
    for items in list1:
        print(items,end= " ")
```

```python
l1 = [10,15,24,75,16]
print(show_list_items(l1))
```

```python
#waf to find the factorial of n
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact *= i
    return fact
```

```python
num = int(input("Enter a number:"))
fact = factorial(num)
print(f"Factorial of {num} is {fact}")
```
