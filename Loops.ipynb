```python
#- Count how many even and odd digits are in a number
old_num = int(input("Enter a number"))
new_num = old_num
odd_num = 0
even_num = 0
while new_num > 0:
    if (new_num % 10) % 2 != 0:
        odd_num = odd_num + 1
    else:
        even_num = even_num + 1
    new_num = new_num // 10

print("Number:",old_num)
print("Total odd:",odd_num)
print("Total even:",even_num)
```

```python
#- Check if a number is a perfect number (sum of divisors equals the number)
num = int(input("Enter a number"))
sum_of_div = 0
for i in range(1,num):
    if num % i == 0:
        sum_of_div = sum_of_div + i

if sum_of_div == num:
    print(num,"is a perfect number.")
else:
    print(num,"is not a perfect number.")
```

```python
#check for a armstrong number
num = int(input("Enter a number:"))
old_num = num
total_digit = 0
while num > 0:
    num = num // 10
    total_digit = total_digit + 1

print(total_digit)
num = old_num
new_num = 0

while num > 0:
    new_num = new_num + (num % 10) ** total_digit
    num = num // 10

if new_num == old_num:
    print(old_num,"is an armstrong number")
else:
    print(old_num,"is not an armstrong number")
```

```python
#count digits and special characters
text = input("Enter a text:")
digit_count = 0
special_count = 0
for char in text:
    if char.isdigit():
        digit_count += 1
    elif not char.isalpha() and not char.isspace():
        special_count += 1
print("Digits:",digit_count)
print("Special characters:",special_count)
```

```python
#frequency of a string using loop and dictionary
text = input("Enter a text:")
freq = {}
for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1
for key in freq:
    print(f"{key}: {freq[key]}")
```
